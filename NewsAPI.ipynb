```python
from newsapi import NewsApiClient
import vaderSentiment.vaderSentiment as vs
import pandas as pd

newsapi = NewsApiClient(api_key="47951fe895b741b597473accded05db7")
analyzer = vs.SentimentIntensityAnalyzer()

top_headlines = newsapi.get_everything(q="Rheinmetall", language="en", sort_by="publishedAt")

articles = top_headlines["articles"]

results = []

for article in articles[:100]:
    title = article["title"]
    score = analyzer.polarity_scores(title)["compound"]

    results.append({
    "Date": article["publishedAt"],
    "Headline": title,
    "Sentiment": score
    })

df_sentiment = pd.DataFrame(results)
avg_live_sentiment = df_sentiment["Sentiment"].mean()

print(f"Analyse abgeschlossen für {len(df_sentiment)} Artikel.")
print(f"Aktuelles Markt-Sentiment für Rheinmetall AG: {avg_live_sentiment: .2f}")

print("\nKritischste Schlagzeilen:")
print(df_sentiment.sort_values(by="Sentiment").head(3))
```
