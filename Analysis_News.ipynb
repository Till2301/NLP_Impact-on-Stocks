```python
import vaderSentiment as vs
analyzer = vs.vaderSentiment.SentimentIntensityAnalyzer()

headlines = [
    "Deutsche Bank shares soar after record profits",
    "Investors worried about rising inflation and bank stability",
    "New partnership expands Deutsche Bank's presence in Asia",
    "Regulators fine major bank over compliance failures"
]

print(f"{'Schlagzeile':<60} | {'Sentiment Score'}")
print("-" * 80)

for h in headlines:
    # VADER gibt ein Dictionary zurück, 'compound' ist der Gesamtwert von -1 bis 1
    score = analyzer.polarity_scores(h)['compound']
    
    status = "NEUTRAL"
    if score >= 0.05: status = "POSITIV"
    elif score <= -0.05: status = "NEGATIV"
    
    print(f"{h:<60} | {score:>15.2f} ({status})")
```
